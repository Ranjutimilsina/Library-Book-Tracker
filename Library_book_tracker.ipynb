```python
#Creating the book class
class Book:
    def __init__(self, title, author, year):
        self.title = title 
        self.author= author 
        self.year = year 
        
    def __str__(self):
        return f" '{self.title}' by '{self.author}' ({self.year})"
        
    
```

```python
# Creating book objects 
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925)
book2 = Book("To Kill a Mockingbird", "Harper Lee", 1960)
book3= Book("Pride and Prejudice", "Jane Austen", 1813)
book4= Book("The Catcher in the Rye", "J.D. Salinger", 1951)
book5= Book("The Kite Runner", "Khaleed Hosseini", 2003)
book6= Book("Becoming", "Michelle Obama", 2018)
```

```python
#Calling book objects 
print(book1)
print(book6)
```

```python
# Defining the library class 
class Library:
    def __init__(self):
        self.books= []
    def add_books(self, book):
        if book not in self.books:
            self.books.append(book)
            print(f"Book '{book.title}'has been added to the library.")
        else:
            print(f"Book '{book.title}' is already in the library.")
    def display_books(self):
        if not self.books:
            print("No books in the library.")
        else:
            print("Books in the library:")
            for book in self.books:
                print(book)
    def search_books(self, keyword):
        found_books=[]
        keyword= keyword.lower()
        for book in self.books:
            if keyword in book.title.lower() or keyword in book.author.lower():
                found_books.append(book)
                
        if found_books:
            print(f"Books found for '{keyword}':")
            for book in found_books:
                print(book)
        else:
            print(f"No books found for '{keyword}'.")
        
    def remove_book(self, title):
        book_to_remove = None
        for book in self.books:
            if book.title.lower() == title.lower():  
                book_to_remove = book
                break
        
        if book_to_remove:
            self.books.remove(book_to_remove)
            print(f"Book '{book_to_remove.title}' has been removed from the library.")
        else:
            print(f"No book found with the title '{title}'.")
            
            
    def display_authors(self):
        if not self.books:
            print("No authors in the library (no books added yet).")
        else:
            unique_authors = {book.author for book in self.books} 
            print("Unique authors in the library:")
            for author in sorted(unique_authors):
                print(author)
                
    def display_sorted_books(self):
        
        if not self.books:
            print("No books in the library.")
        else:
            print("Books in the library (sorted by title):")
            sorted_books = sorted(self.books, key=lambda book: book.title)
            for book in sorted_books:
                print(book)
                
                
    def __init__(self):
        self.books = []
        self.borrowed_books = []  
    def borrow_book(self, title, borrower):
        for book in self.books:
            if book.title.lower() == title.lower():  
                self.books.remove(book)
                self.borrowed_books.append((book, borrower))  
                print(f"Book '{book.title}' has been borrowed by {borrower}.")
                return
        print(f"Book '{title}' is not available in the library.")
    
    def display_borrowed_books(self):
        if not self.borrowed_books:
            print("No books are currently borrowed.")
        else:
            print("Borrowed books:")
            for book, borrower in self.borrowed_books:
                print(f"'{book.title}' by {book.author}, borrowed by {borrower}")
                
                
    def return_book(self, title):
        for book, borrower in self.borrowed_books:
            if book.title.lower() == title.lower():
                self.borrowed_books.remove((book, borrower))
                self.books.append(book)
                print(f"Book '{book.title}' has been returned to the library.")
                break
        else:
            print(f"No borrowed book found with the title '{title}'.")


```

```python
#Creating library 
library = Library()

#Adding book to the library 
library.add_books(book1)
library.add_books(book2)
library.add_books(book3)
library.add_books(book4)
library.add_books(book5)
library.add_books(book6)


#Checking by adding a duplicate 
library.add_books(book1)

# Display books in the library 
library.display_books()

# Searching books by a keyword
library.search_books("gatsby")
library.search_books("harper")  
library.search_books("harry")  

# Removing a book by title
library.remove_book("the kite runner") 
library.remove_book("Moby Dick") 

# Display all books after removal
library.display_books()

# Display unique authors
library.display_authors()

# Display books sorted by title
library.display_sorted_books()

# Borrowing a book- example
library.borrow_book("becoming", "Alice")

# Display borrowed books
library.display_borrowed_books()

# Returning a book
library.return_book("becoming")
library.return_book("to kill a mocking bird")

```
